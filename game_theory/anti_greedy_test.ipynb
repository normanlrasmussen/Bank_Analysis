# This file shows the effectivness of the anit-greedy vs the greedy and sesua greedy player

```python
from bank import Bank
from players import GreedyPlayer, SesquaGreedyPlayer, AntiGreedyPlayer, ThersholdPlayer
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Test aganist GreedyPlayer
players = [GreedyPlayer(), AntiGreedyPlayer()]
wins, ties = Bank.estimate_win_probability(players, 10, num_simulations=10000)
print(f"Anti Greedy win: {wins[1]}, Greedy win: {wins[0]}, Tie: {ties[0]}")
```

```python
# Test against SesquaGreedyPlayer
players = [SesquaGreedyPlayer(), AntiGreedyPlayer()]
wins, ties = Bank.estimate_win_probability(players, 10, num_simulations=10000)
print(f"Anti Greedy win: {wins[1]}, SesquaGreedy win: {wins[0]}, Tie: {ties[0]}")
```

```python
# Test against SesquaGreedyPlayer
players = [SesquaGreedyPlayer(), AntiGreedyPlayer()]
wins, ties = Bank.estimate_win_probability(players, 1, num_simulations=10000)
print(f"Anti Greedy win: {wins[1]}, SesquaGreedy win: {wins[0]}, Tie: {ties[0]}")
```

```python
# Plot the effect of rounds on AntiGreedyPlayer win rate
number_of_rounds = [i for i in range(1, 51, 2)]  # Test rounds from 1 to 50

anti_greedy_vs_greedy_win = []
anti_greedy_vs_sesqua_win = []
anti_greedy_vs_greedy_tie = []
anti_greedy_vs_sesqua_tie = []

print("Testing different numbers of rounds...")
for rounds in number_of_rounds:
    print(f"Testing {rounds} rounds...", end=" ")
    
    # Test against GreedyPlayer
    players = [GreedyPlayer(), AntiGreedyPlayer()]
    wins, ties = Bank.estimate_win_probability(players, rounds, num_simulations=5000)
    anti_greedy_vs_greedy_win.append(wins[1])
    anti_greedy_vs_greedy_tie.append(ties[0])
    
    # Test against SesquaGreedyPlayer
    players = [SesquaGreedyPlayer(), AntiGreedyPlayer()]
    wins, ties = Bank.estimate_win_probability(players, rounds, num_simulations=5000)
    anti_greedy_vs_sesqua_win.append(wins[1])
    anti_greedy_vs_sesqua_tie.append(ties[0])
    
    print("Done")

print("\nSimulation complete!")
```

```python
# Plot the results
fig, ax = plt.subplots(figsize=(12, 7))

# Plot win rates
ax.plot(number_of_rounds, anti_greedy_vs_greedy_win, label="AntiGreedy Win % (vs Greedy)", 
        marker='o', color='green', linewidth=2, markersize=4)

ax.set_xlabel("Number of Rounds", fontsize=12)
ax.set_ylabel("Percentage", fontsize=12)
ax.set_title("Effect of Rounds on AntiGreedyPlayer Win Rate", fontsize=14, fontweight='bold')
ax.legend(loc='best', fontsize=10)
ax.grid(True, alpha=0.3)
ax.axhline(0.5, color='gray', linestyle='--', linewidth=1)
ax.set_ylim([0, 1])
plt.savefig("anti_greedy_test_rounds.png")

plt.tight_layout()
plt.show()
```

```python
# Plot the results
fig, ax = plt.subplots(figsize=(12, 7))

# Plot win rates
ax.plot(number_of_rounds, anti_greedy_vs_greedy_win, label="AntiGreedy Win % (vs Threshold)", 
        marker='o', color='green', linewidth=2, markersize=4)

ax.set_xlabel("Number of Rounds", fontsize=12)
ax.set_ylabel("Percentage", fontsize=12)
ax.set_title("Effect of Rounds on AntiGreedyPlayer Win Rate", fontsize=14, fontweight='bold')
ax.legend(loc='best', fontsize=10)
ax.grid(True, alpha=0.3)
ax.set_ylim([0, 1])
ax.axhline(0.5, color='gray', linestyle='--', linewidth=1)
plt.savefig("anti_greedy_test_rounds_1.png")

plt.tight_layout()
plt.show()
```
