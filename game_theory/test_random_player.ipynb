```python
from bank import Bank
from players import ThersholdPlayer, GreedyPlayer, ProbabilisticPlayer
```

```python
probs = [1/10, 1/9, 1/8, 1/7, 1/6, 1/3, 1/2, 2/3, 5/6]
players = [ProbabilisticPlayer(probability=p) for p in probs]

expected_values = Bank.get_expected_value(players, 10, 10000)

for i in range(len(probs)):
    print(f"Probability: {probs[i]:.2f}, Expected Value: {expected_values[i]:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar([f"{p:.2f}" for p in probs], expected_values)
plt.xlabel("Bank Probability")
plt.ylabel("Expected Value")
plt.title("Expected Value vs. Banking Probability (ProbabilisticPlayer)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Play a ProbabilisticPlayer versus a ThersholdPlayer
prob_player = ProbabilisticPlayer(name="Prob", probability=1/10)
thresh_player = ThersholdPlayer(name="Thresh", threshold=200)

players = [prob_player, thresh_player]

# Simulate and print expected values and win probability for each player
expected_values, win_prob, tie_prob = Bank.get_all(players, rounds=10, num_simulations=10000)

print("Expected Value for each player (ProbabilisticPlayer, ThersholdPlayer):")
for i, player in enumerate(players):
    print(f"{player.name}: {expected_values[i]:.2f}")

print("\nProbability of Winning:")
for i, player in enumerate(players):
    print(f"{player.name}: {win_prob[i]*100:.2f}%")

print("\nProbability of Tie (i.e., both players tie):")
print(f"Tie Probability: {tie_prob[0]*100:.2f}%")


```
