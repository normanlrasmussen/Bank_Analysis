```python
import numpy as np
from bank import Bank
from players import ThersholdPlayer
import matplotlib.pyplot as plt
```

# Show that threshold 100 is the better than 2500

```python
players = [ThersholdPlayer(threshold=100), ThersholdPlayer(threshold=2500)]

win_probs, tie_probs = Bank.estimate_win_probability(players, rounds=10, num_simulations=10000)

# Plot win and tie probabilities stacked, with labels

player_labels = [f"Player {i}" for i in range(len(players))]

# Stacked bar chart: win below, tie above
fig, ax = plt.subplots(figsize=(7, 5))
bars1 = ax.bar(player_labels, win_probs, label="Win Probability")
bars2 = ax.bar(player_labels, tie_probs, bottom=win_probs, label="Tie Probability")

ax.set_ylabel("Probability")
ax.set_title("Win and Tie Probabilities per Player")
ax.legend()

for i, (w, t) in enumerate(zip(win_probs, tie_probs)):
    ax.text(i, w/2, f"{w:.2%}", ha='center', va='center', color='white', fontsize=10, weight='bold')
    ax.text(i, w + t/2, f"{t:.2%}", ha='center', va='center', color='white', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()

```
