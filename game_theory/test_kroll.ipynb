```python
from bank import Bank
from players import RollThresholdPlayer
import matplotlib.pyplot as plt
```

```python
rolls = [i for i in range(1, 101)]
players = [RollThresholdPlayer(threshold=roll) for roll in rolls]

expected_values = Bank.get_expected_value(players, 10, 50000)

for i in range(len(rolls)):
    print(f"Threshold: {rolls[i]}, Expected Value: {expected_values[i]:.2f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(rolls, expected_values)
plt.xlabel("Amount of Rolls")
plt.ylabel("Expected Value")
plt.title("Expected Value vs. Amount of Rolls")
plt.grid()
plt.show()
```
