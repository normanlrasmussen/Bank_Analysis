```python
import numpy as np
import matplotlib.pyplot as plt

from players import GreedyPlayerK, GreedyPlayer, ThersholdPlayer
from bank import Bank
```

```python
threshold  =75
```

```python
win_percentages = []
tie_percentages = []
ks = list(range(1, 21))
rounds = 10
num_simulations = 1000

# Do for regular greedy
players = [GreedyPlayer(), ThersholdPlayer(threshold=threshold)]
_, win_percentage, tie_percentage = Bank.get_all(players, rounds=rounds, num_simulations=num_simulations)
win_percentages.append(win_percentage[0])
tie_percentages.append(tie_percentage[0])

# Do for greedyk
for k in ks:
    players = [ThersholdPlayer(threshold=threshold), GreedyPlayerK(k=k)]
    _, win_percentage, tie_percentage = Bank.get_all(players, rounds=rounds, num_simulations=num_simulations)
    # win_percentage[1] is GreedyPlayerK, tie_percentage[1] is GreedyPlayerK's tie rate
    win_percentages.append(win_percentage[1])
    tie_percentages.append(tie_percentage[1])

plt.figure(figsize=(8, 6))
plt.plot([0] + ks, win_percentages, marker='o', label="GreedyPlayerK Win %")
plt.plot([0] + ks, tie_percentages, marker='s', label="GreedyPlayerK Tie %")
plt.xlabel("k value (GreedyPlayerK)")
plt.ylabel("Percentage")
plt.title("Win and Tie Percentages of GreedyPlayerK vs ThresholdPlayer")
plt.legend()
plt.grid(True)
plt.show()
```
