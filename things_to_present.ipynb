```python
from bank import Bank
from player import ThersholdPlayer, GreedyPlayer
import matplotlib.pyplot as plt
```

# Weird stuff keeps happening as you increase the threshold

```python

# thresholds = [50 + 500 * i for i in range(100)]
thresholds = [50 + 10 * i for i in range(1000)]
# thresholds = [50, 844]
players = [ThersholdPlayer(threshold=threshold) for threshold in thresholds]

expected_score, winner_count, tie_count = Bank.get_all(players, 10, num_simulations=10000)

# for val, threshold, winner_prob in zip(expected_score, thresholds, winner_count):
#     print(f"Threshold: {threshold}, Expected Score: {val}, Winner Probability: {winner_prob}")



# Plot the expected score and winner probability vs. threshold

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Threshold')
ax1.set_ylabel('Expected Score', color=color)
ax1.plot(thresholds, expected_score, marker='o', color=color, label='Expected Score')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:orange'
ax2.set_ylabel('Winner Probability', color=color)
ax2.plot(thresholds, winner_count, marker='s', color=color, label='Winner Probability')
ax2.tick_params(axis='y', labelcolor=color)
# Plot the tie probability along with winner probability

color_tie = 'tab:green'
ax2.plot(thresholds, tie_count, marker='^', color=color_tie, label='Tie Probability')

# Show both lines' legends on ax2 (right y-axis)
lines, labels = ax2.get_legend_handles_labels()
lines1, labels1 = ax1.get_legend_handles_labels()
ax2.legend(lines + lines1, labels + labels1, loc='upper right')

plt.title("Expected Score and Winner Probability vs. Threshold")
fig.tight_layout()
plt.grid(True)
plt.show()
```

```python
# Calculate non-loss probability (win + tie) for each threshold
non_loss_probabilities = [(w + t) for w, t in zip(winner_count, tie_count)]

plt.figure(figsize=(8, 5))
plt.plot(thresholds, [p * 100 for p in non_loss_probabilities], marker='o', color='purple', label='Non-loss Probability (%)')
plt.xlabel('Threshold')
plt.ylabel('Non-loss Probability (%)')
plt.title('Non-loss Probability vs. Threshold')
plt.grid(True)
plt.legend()
plt.show()
```

# The greedy player will win at 2 player games

```python
thresholds = [50 + 50 * i for i in range(50)]

win_data = []

for threshold in thresholds:
    players = [GreedyPlayer(), ThersholdPlayer(threshold=threshold)]
    winner_count, tie_count = Bank.estimate_win_probability(players, 10, num_simulations=1000)
    print(f"Threshold: {threshold}")
    print(f"Greedy Wins: {winner_count[0]}")
    print(f"Threshold Wins: {winner_count[1]}")
    print(f"Tie: {tie_count[0]}")
    print(f" ")

    win_data.append((winner_count[0], winner_count[1], tie_count[0]))
```

```python
# Calculate greedy win, tie, and non-loss percentages for each threshold
threshold_win_percentages = [win[1] * 100 for win in win_data]
greedy_win_percentages = [win[0] * 100 for win in win_data]
tie_percentages = [win[2] * 100 for win in win_data]
non_loss_percentages = [g + t for g, t in zip(greedy_win_percentages, tie_percentages)]

plt.figure(figsize=(8, 5))
plt.plot(thresholds, greedy_win_percentages, marker='o', color='blue', label='Greedy Win %')
plt.plot(thresholds, tie_percentages, marker='s', color='orange', label='Tie %')
plt.plot(thresholds, non_loss_percentages, marker='^', color='green', label='Non-loss % (Win + Tie)')
plt.plot(thresholds, threshold_win_percentages, marker='x', color='red', label='Threshold Win %')

plt.xlabel('Threshold')
plt.ylabel('Percentage')
plt.title('Greedy Player Win, Tie, and Non-loss Percentage vs Threshold')
plt.ylim(0, 100)
plt.grid(True)
plt.legend()
plt.show()
```
